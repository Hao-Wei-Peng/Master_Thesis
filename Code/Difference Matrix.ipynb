```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('seaborn-whitegrid')
import seaborn as sns
```

```python
Baseline_CIFAR10_confusion_matrix = np.load("Baseline_CIFAR10_Confusion_Matrix.npy")
Curriculum_CIFAR10_confusion_matrix = np.load("Curriculum_CIFAR10_Confusion_Matrix.npy")

Diff_matrix_10 = Curriculum_CIFAR10_confusion_matrix - Baseline_CIFAR10_confusion_matrix

import seaborn as sn

label = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]
chart_label = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]

# Change confusion matrix type to pd dataframe
Diff_matrix_10 = pd.DataFrame(Diff_matrix_10)

# Plot the CIFAR10 heatmap confusion matrix 
ax = plt.subplots(figsize = (10, 8)) 
ax = plt.axes()
sn.heatmap(Diff_matrix_10, ax = ax, annot = True, cmap = "Purples", fmt = "g")

# plt.xticks(np.arange(len(chart_label)), chart_label, rotation = 90, ha = "left")
# plt.yticks(np.arange(len(label)), label, rotation = 180, ha = "left")

ax.set_xticklabels(chart_label, rotation = 90)
ax.set_yticklabels(label, rotation = 360)


ax.set_title("In Domain (CIFAR10) Difference Results", fontweight = "bold")
plt.xlabel("Predict", fontweight = "bold")
plt.ylabel("Answer", fontweight = "bold")

plt.show()
```

```python
Baseline_CIFAR100_confusion_matrix = np.load("Baseline_CIFAR100_Confusion_Matrix.npy")
Curriculum_CIFAR100_confusion_matrix = np.load("Curriculum_CIFAR100_Confusion_Matrix.npy")

Diff_matrix = Curriculum_CIFAR100_confusion_matrix - Baseline_CIFAR100_confusion_matrix

import seaborn as sn

label = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]
chart_label = ["rocket", "pickup truck", "butterfly", "leopard", "cattle", "wolf", "lizard", "camel", "tank", "bus"]

# Change confusion matrix type to pd dataframe
Diff_matrix = pd.DataFrame(Diff_matrix)

# Plot the CIFAR10 heatmap confusion matrix 
ax = plt.subplots(figsize = (10, 8)) 
ax = plt.axes()
sn.heatmap(Diff_matrix, ax = ax, annot = True, cmap = "Greys", fmt = "g")

# plt.xticks(np.arange(len(chart_label)), chart_label, rotation = 90, ha = "left")
# plt.yticks(np.arange(len(label)), label, rotation = 180, ha = "left")

ax.set_xticklabels(chart_label, rotation = 90)
ax.set_yticklabels(label, rotation = 360)


ax.set_title("Out of Domain (CIFAR100) Difference Results", fontweight = "bold")
plt.xlabel("Predict", fontweight = "bold")
plt.ylabel("Answer", fontweight = "bold")

plt.show()
```
